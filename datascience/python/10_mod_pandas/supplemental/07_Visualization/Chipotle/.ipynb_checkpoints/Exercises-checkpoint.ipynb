# Visualizing Chipotle's Data

This time we are going to pull data directly from the internet.
Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
x = chipo['item_name'].value_counts().head()
x.plot(kind='bar')
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')
plt.show
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price]
y = chipo.groupby('order_id').sum()

plt.scatter(x = y['item_price'], y = y['quantity'], s = 50, c = 'green')
plt.xlabel('Price')
plt.ylabel('Items')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```

### Step 7. BONUS: Create a question and a graph to answer your own question. What was the least ordered item?

```python
x = chipo['item_name'].value_counts().tail()
x.plot(kind='bar')
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Least ordered Chipotle\'s Items')
plt.show
```
